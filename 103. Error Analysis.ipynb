### Load Dataset

```python
import pandas as pd 
df = pd.read_csv(raw_data_path, encoding = 'unicode_escape')
test_df = pd.read_csv(result_file_path)
```

### Dataset Set up

```python
df = df.melt(id_vars=["Domain", "Sub domain", "Intent"]).drop("variable", axis = 1)
Classes_dict_1 = dict(zip(list(df["Intent"].unique()), [i for i in range(df["Intent"].nunique())]))
Classes_dict = {value:key for key, value in Classes_dict_1.items()}
test_df["y_true"] = test_df.Intent.apply(lambda x: Classes_dict_1.get(x, "Unknow"))
test_df["model_correctness"] = test_df["y_true"]  == test_df["y_hat"]
test_df["y_hat"] = test_df.y_hat.apply(lambda x: Classes_dict.get(x, "Unknow"))
```

```python
df_wrong = test_df.loc[test_df.model_correctness == False]
df_wrong[df_wrong.y_hat != "Unknow"]
```

```python
print("The testing accuracy is:", test_df.model_correctness.mean())
```
